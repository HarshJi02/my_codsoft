# IMORTING LIBRARIES

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('creditcard.csv',sep=',')
data.head()
```

```python
data.info()
```

```python
data.isnull().values.any()
```

```python
data['Class'].value_counts()
```

```python
# separating data for analysis
legit = data[data.Class == 0]  
fraud = data[data.Class == 1]  
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatenating two dataframes
new = pd.concat([legit_sample, fraud], axis=0)
new.head()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
x = new.drop(columns='Class', axis=1)
y = new['Class']
print(x)
```

```python
print(y)
```

# SPLITTING DATA

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

# MODEL TRAINING

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(train_prediction, y_train)
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(test_prediction, y_test)
print('Accuracy score on Test data: ', test_data_accuracy)
```
